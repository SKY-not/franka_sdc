```python
import cv2
cap = cv2.VideoCapture(1) # 0代表第一个摄像头，如有多个可尝试1,2...
# 设置分辨率（推荐1280x720或1920x1080，平衡清晰度和速度）
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 1280)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 720)
# 设置帧率（例如30帧/秒）
cap.set(cv2.CAP_PROP_FPS, 0)
# 尝试调整自动曝光等（可选，部分相机支持）
# cap.set(cv2.CAP_PROP_AUTO_EXPOSURE, 0.25) # 关闭自动曝光
# cap.set(cv2.CAP_PROP_EXPOSURE, -4) # 手动设置曝光值
while True:
    ret, frame = cap.read()
    if ret:
        # 此处处理frame（图像帧），例如进行目标识别
        cv2.imshow('Frame', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
